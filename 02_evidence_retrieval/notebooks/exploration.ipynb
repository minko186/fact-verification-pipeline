```python
import datasets
import nltk
from sentence_transformers import SentenceTransformer
import chromadb
from chromadb.utils import embedding_functions

# Download NLTK sentence tokenizer
nltk.download("punkt")

# 1. Load Wikipedia dataset from Hugging Face
dataset = datasets.load_dataset("wikimedia/wikipedia", "20231101.en", split="train[:1%]")  
# Using 1% for demo – full dump is huge!

# 2. Split articles into sentences
def split_sentences(example):
    sentences = nltk.sent_tokenize(example["text"])
    return {"sentences": sentences}

dataset = dataset.map(split_sentences, batched=True, remove_columns=["text"])
all_sentences = sum(dataset["sentences"], [])  # flatten

print(f"Total sentences: {len(all_sentences)}")

# 3. Setup embedding model
model = SentenceTransformer("sentence-transformers/all-mpnet-base-v2")

# 4. Setup Chroma DB
client = chromadb.Client()
collection = client.create_collection(
    name="wikipedia_facts",
    embedding_function=embedding_functions.SentenceTransformerEmbeddingFunction(model_name="sentence-transformers/all-mpnet-base-v2")
)

# Add to Chroma
batch_size = 1000
for i in range(0, len(all_sentences), batch_size):
    batch = all_sentences[i:i+batch_size]
    ids = [f"sent_{i+j}" for j in range(len(batch))]
    collection.add(documents=batch, ids=ids)

print("Knowledge base created ✅")

# 5. Querying with a fact
def query_fact(fact, top_k=5):
    results = collection.query(
        query_texts=[fact],
        n_results=top_k
    )
    return results["documents"][0]

# Example usage
fact = "Albert Einstein developed the theory of relativity."
evidence = query_fact(fact, top_k=5)

print("\nQuery:", fact)
print("Evidence:")
for e in evidence:
    print("-", e)
```
