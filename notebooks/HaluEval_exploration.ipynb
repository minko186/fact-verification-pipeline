```python
import json
from datasets import Dataset

# -------------------------------
# CONFIG
# -------------------------------
FILES = {
    "minko186/halueval-dialogue": "dialogue_data.json",
    "minko186/halueval-general": "general_data.json",
    "minko186/halueval-qa": "qa_data.json",
    "minko186/halueval-summarization": "summarization_data.json",
}

# -------------------------------
# FLEXIBLE JSON LOADER
# -------------------------------
def load_json_flex(path):
    """Handles JSONL or multi-object JSON files safely."""
    with open(path, "r", encoding="utf-8") as f:
        text = f.read().strip()
        try:
            return json.loads(text)  # standard JSON array
        except json.JSONDecodeError:
            # fallback: JSONL (one object per line)
            return [json.loads(line) for line in text.splitlines() if line.strip()]

# -------------------------------
# LOAD + PUSH EACH DATASET
# -------------------------------
for repo_id, file_path in FILES.items():
    print(f"\n🔹 Loading {file_path} ...")
    data = load_json_flex(file_path)
    print(f"✅ Loaded {len(data):,} records")

    dataset = Dataset.from_list(data)

    print(f"🚀 Pushing to Hugging Face Hub: {repo_id} (public)")
    dataset.push_to_hub(repo_id, private=False)
    print(f"✅ Uploaded successfully: https://huggingface.co/datasets/{repo_id}")

print("\n🎉 All HaluEval datasets pushed successfully!")
```
