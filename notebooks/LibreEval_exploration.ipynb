```python
import pandas as pd
from datasets import Dataset

# -------------------------------
# CONFIG
# -------------------------------
FILES = {
    "minko186/libreeval-synthetic-hallucinations-en": "synthetic_hallucinations_english.csv",
    "minko186/libreeval-non-synthetic-hallucinations-en": "non_synthetic_hallucinations_english.csv",
}

# -------------------------------
# LOAD + PUSH EACH CSV
# -------------------------------
for repo_id, file_path in FILES.items():
    print(f"\n🔹 Loading {file_path} ...")
    df = pd.read_csv(file_path)
    print(f"✅ Loaded {len(df):,} rows with columns: {list(df.columns)}")

    dataset = Dataset.from_pandas(df)

    print(f"🚀 Pushing to Hugging Face Hub: {repo_id} (public)")
    dataset.push_to_hub(repo_id, private=False)
    print(f"✅ Uploaded successfully: https://huggingface.co/datasets/{repo_id}")

print("\n🎉 All datasets pushed successfully!")
```
